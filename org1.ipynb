### 4. 결측치 확인
컬럼별 결측치(NaN) 개수를 확인하고 필터링해 봅니다.

```python
df.info()
```

```python
# 1. 전체 컬럼별 결측치 개수
print("--- 전체 결측치 개수 ---")
print(df.isnull().sum())

# 2. 결측치가 있는 컬럼과 개수만 쏙 골라 출력
print("\n--- 결측치가 있는 컬럼만 필터링 ---")
null_counts = df.isnull().sum()
print(null_counts[null_counts > 0])
```

```python
null_counts = df.isnull().sum()
```

```python
null_counts[null_counts > 0]
```

```python
median_month = df['MonthlyIncome'].median()
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(median_month)
df
```

```python
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(0)
df
```

```python
df = df[df['MonthlyIncome'] <= 30000]
df
```

```python
df
```

```python
df = df.drop(columns=['Employee_ID'])
df
```

```python
import seaborn as sns

# x축: 나이(Age), y축: 월소득(MonthlyIncome), 색상구분(hue): 퇴사여부(Attrition)
sns.scatterplot(data=df, x='Age', y='MonthlyIncome', hue='Attrition')

# 시각화 화면 출력
plt.show()
```

```python
df = pd.get_dummies(df, columns=['Gender', 'OverTime', 'Department'], drop_first=True)
```
